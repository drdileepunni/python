```python
import pandas as pd
import numpy as np
```

```python
nn = pd.read_csv('NN_source_file.csv', index_col=[0])
```

```python
nn.shape
```

```python
nn.info()
```

```python
nn['Temparature'] = pd.to_numeric(nn['Temparature'], errors='coerce') 
```

```python
nn.Temparature = nn.Temparature.replace(np.nan, 98.6)
```

```python
import numpy as np
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(scaler.fit_transform(pd.DataFrame(df[col])))
    return df
    
```

```python
nn['Month.of.Admission'].unique()
```

```python
nn['Month.of.Admission'] = nn['Month.of.Admission'].replace('July 0201', 'July 2018')
nn['Month.of.Admission'] = nn['Month.of.Admission'].replace('January 2018', 'January 2019')
nn['Month.of.Admission'] = nn['Month.of.Admission'].replace('August 2015', 'August 2018')
nn['Month.of.Admission'] = nn['Month.of.Admission'].replace('February 2018', 'February 2019')
nn['Month.of.Admission'] = nn['Month.of.Admission'].replace('February 0219', 'February 2019')
```

```python
nn['Vent.mode'].value_counts()
```

```python
nn['Vent.mode'] = nn['Vent.mode'].replace(np.nan, 'RA')
```

```python
nn['Gender'].unique()
```

```python
nn = nn.drop(['Name', 'CPMRN', 'History', 'SBP'], axis=1)
```

```python
nn.isnull().sum()
```

```python
nn_scaled = scaleColumns(nn, ['GCS', 'FiO2', 'PaO2', 'MAP', 'Age', 'RR', 'HR', 'pH', 'TLC',
       'Serum.Cr', 'Na', 'Temparature', 'Hb'])
```

```python
nn_scaled
```

```python
fin = pd.get_dummies(nn_scaled, drop_first=True)
```

```python
fin = fin.dropna()
fin.isnull().sum().sum()
```

```python
X = fin.iloc[:,0:47].values
```

```python
y = fin.iloc[:,47].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, stratify=y)
```

```python
model = Sequential()
model.add(Dense(100, activation='relu', input_shape=(47,)))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_curve
y_pred_prob = model.predict(X_train)
tpr, fpr, thresholds = roc_curve(y_train, y_pred_prob)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score

roc_auc_score(y_train, y_pred_prob)
```

```python
from sklearn.metrics import roc_curve
y_pred_prob = model.predict(X_train)
tpr, fpr, thresholds = roc_curve(y_train, y_pred_prob)
```
