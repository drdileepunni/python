```python
anal = pd.read_csv('/home/dileep/Desktop/python/All_csvs/daily_note.csv')
indices = np.random.random_integers(0,1460,100)
sample = anal.iloc[indices, :]
sample_notes = sample.iloc[:,np.arange(58,86)]
sample_notes = sample_notes.drop(sample_notes.columns[np.arange(8,28)], axis=1)
```

```python
d0 = sample_notes['Day0']
```

```python
from quickumls import QuickUMLS
matcher = QuickUMLS(quickumls_fp='/home/dileep/QuickUMLS', threshold=0.7)
```

```python
terms = []
for idx, note in enumerate(d0):
    terms.append(matcher.match(note, best_match=True, ignore_syntax=False))
```

```python
medicine_set = {"T116","T087","T195","T123","T026","T103","T120","T104","T200","T088","T203","T196","T126","T125",
                "T197","T114","T086","T109","T121","T167","T127"}
disease_set = {"T020","T190","T007","T019","T049","T047","T004","T131","T037","T048","T191","T005", "T039"}
finding_set = {"T017","T022","T053","T038","T029","T201","T033","T041","T070","T042","T046","T184"}
lab_set = {"T031","T130","T059","T034"}
procedure_set = {"T091","T122","T023","T030","T060","T021","T074","T061"}
```

```python
# Finding Disease processes
for i in range(len(terms[3])):
    for j in range(len(terms[3][i])):
        if (bool(terms[3][i][j]['semtypes'].intersection(disease_set))):
            print('Disease processes:{}'.format(terms[3][i][j]['ngram']))
        if (bool(terms[3][i][j]['semtypes'].intersection(medicine_set))):
            print('Medicines: {}'.format(terms[3][i][j]['ngram']))
        if (bool(terms[3][i][j]['semtypes'].intersection(finding_set))):
            print('Examination: {}'.format(terms[3][i][j]['ngram']))
```
