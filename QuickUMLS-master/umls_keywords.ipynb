```python
import json
with open('popular_patients.json') as json_file:
    all_pats = json.load(json_file)
```

```python
notes = []
for i in range(len(all_pats)):
    pat_obj = all_pats[i]
    if (pat_obj.get('notes')):
        summary = []; ap = []; to_do = []; dx = []
        summary.append(pat_obj['notes']['finalNotes'][0]['content'][0]['primaryText'])
        ap.append(pat_obj['notes']['finalNotes'][0]['content'][0]['secondaryText'])
        to_do.append(pat_obj['notes']['finalNotes'][0]['content'][0]['tertiaryText'])
        notes.append([summary, ap, to_do, dx])
```

```python
medicine_set = {"T116","T087","T195","T123","T026","T103","T120","T104","T200","T088","T203","T196","T126","T125",
                "T197","T114","T086","T109","T121","T167","T127"}
disease_set = {"T020","T190","T007","T019","T049","T047","T004","T131","T037","T048","T191","T005"}
exam_finding_set = {"T017","T022","T053","T038","T029","T201","T033","T041","T070","T042","T046","T184"}
lab_related_set = {"T031","T130","T059","T034"}
procedure_related_set = {"T091","T122","T023","T030","T060","T021","T074","T061"}
```

```python
from quickumls import QuickUMLS
matcher = QuickUMLS(quickumls_fp='/home/dileep/QuickUMLS', threshold=0.7)
```

```python
def getUMLSTerms(note):
    a = matcher.match(note, best_match=True, ignore_syntax=False)
    # i is the index of list of patient notes
    # j is the index of terms in each patient note
    # k is the index of matches in each term-match list
    perfect_matches = []
    match_pat = []
    for j in range(len(a)):
        for k in range(len(a[j])):
            if (a[j][k]['similarity'] >= 0.8):
                match_pat.append([a[j][k]['ngram'], a[j][k]['semtypes']])
        perfect_matches.append(match_pat)
    print('\ndiagnoses...')
    for i in range(len(perfect_matches[2])):
        if (bool(perfect_matches[2][i][1].intersection(disease_set))):
            print(perfect_matches[2][i][0])
    print('\nexam findings...')
    for i in range(len(perfect_matches[2])):
        if (bool(perfect_matches[2][i][1].intersection(exam_finding_set))):
            print(perfect_matches[2][i][0])
    # Finding Medicines
    #print('\nmedicines...')
    #for i in range(len(perfect_matches[2])):
    #    if (bool(perfect_matches[2][i][1].intersection(medicine_set))):
            print(perfect_matches[2][i][0])
    # Finding Disease processes
    # Finding Examination findings
    
    # Finding Labs findings
    #print('\nlabs...')
    #for i in range(len(perfect_matches[2])):
    #    if (bool(perfect_matches[2][i][1].intersection(lab_related_set))):
    #        print(perfect_matches[2][i][0])
    # Finding Procedural stuff
    #print('\nprocedures...')
    #for i in range(len(perfect_matches[2])):
    #    if (bool(perfect_matches[2][i][1].intersection(procedure_related_set))):
    #        print(perfect_matches[2][i][0])
```

```python
for i in range(len(notes)):
    my_note = notes[i][0]
    print(my_note)
    getUMLSTerms(my_note)
    print('----------------------------------------------------')
```

```python
match_list = []
for i in range(len(perfect_matches)):
    patient = []
    med = []; disease = []; exam = []; lab = []; procedure = [];
    for j in range(len(perfect_matches[i])):
        if (bool(perfect_matches[i][j][1].intersection(medicine_set))):
            med.append(perfect_matches[i][j][0])
        if (bool(perfect_matches[i][j][1].intersection(disease_set))):
            disease.append(perfect_matches[i][j][0])
        if (bool(perfect_matches[i][j][1].intersection(exam_finding_set))):
            exam.append(perfect_matches[i][j][0])
        if (bool(perfect_matches[i][j][1].intersection(lab_related_set))):
            lab.append(perfect_matches[i][j][0])
        if (bool(perfect_matches[i][j][1].intersection(procedure_related_set))):
            procedure.append(perfect_matches[i][j][0])
    patient.append(med)
    patient.append(disease)
    patient.append(exam)
    patient.append(lab)
    patient.append(procedure)
    match_list.append(patient)
```
